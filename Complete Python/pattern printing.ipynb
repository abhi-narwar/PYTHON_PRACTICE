```python
'''loop inside a loop
Practice Questions:

Print star pattern.
Print number pattern.
Print square pattern.
Print multiplication tables from 1 to 5.
Print triangle pattern.'''
```

```python
n = 5
for i in range(1, n + 1):
    for j in range(1, n + 1):
        if j <= i:
            print(j, end="")
    print()
```

```python
n = 5
for i in range(1, n + 1):
    for j in range(1, n + 1):
        if j <= i:
            print('*', end="")
    print()
```

```python
n = 5
for i in range(n):
    print("*" * n)
```

```python
for i in range(1, 6):
    print(f"\nTable of {i}")
    for j in range(1, 11):
        print(f"{i} x {j} = {i*j}")
```

```python
n = 5
for i in range(1, n + 1):
    print(" " * (n - i) + "*" * i)
```

```python
'''â­ Interview Important Programs
Factorial of a number
Prime number check
Fibonacci series
Count vowels in a string
Sum of even and odd numbers separately'''
```

```python
n=5
fact=1
while n>0:
    fact=fact*n
    n-=1
print(fact)
```

```python
def prime(n):
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1

    if count == 2:
        print("prime number")
    else:
        print("not a prime number")

n = 5
prime(n)
```

```python
s = "abhishek"
vowels = "aeiouAEIOU"
count = 0

for ch in s:
    if ch in vowels:
        count += 1

print("Vowel count:", count)
```

```python
numbers = [1, 2, 3, 4, 5, 6]
even_sum = 0
odd_sum = 0

for num in numbers:
    if num % 2 == 0:
        even_sum += num
    else:
        odd_sum += num

print("Sum of even numbers:", even_sum)
print("Sum of odd numbers:", odd_sum)
```

```python
#fibonacci series
n = 7
a, b = 0, 1

for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
```
