# ðŸ”¹ while Loop

```python
'''Used when the number of iterations is not known.

Practice Questions:
Print numbers from 10 to 1.
Find sum of digits of a number.
Reverse a number.
Count digits in a number.
Check whether a number is palindrome.'''
```

```python
n=10
while n>0:
    print(n,end=" ")
    n-=1
    
```

```python
digit_sum=0
n=1234
while n>0:
    r=n%10
    digit_sum+=r
    n=n//10
print(digit_sum)
```

```python
#reverse

n=1234
rev=0
while n>0:
    r=n%10
    rev=rev*10+r
    n=n//10
print(rev)
```

```python
#digit count


n=242553
count=0
while n>0:
    r=n%10
    count+=1
    n=n//10
print(count)
    
```

```python
#palindrome or not

n=121
temp=n
rev=0
while n>0:
    r=n%10
    rev=rev*10+r
    n=n//10
if ==rev:
    print("palindrome")
else:
    print("not a palindrome")
    
```
